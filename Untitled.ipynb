# Describing Distributions



```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
# Loading in data

df = pd.read_csv('Data/Medical_Dataset.csv')
df.head()
```

```python
df['VitD_levels'].isna().sum()
```

```python
# getting mean/median/std for plotting
vitD_mean = df['VitD_levels'].mean()
vitD_median = df['VitD_levels'].median()
vitD_std = df['VitD_levels'].std()
vitD_std_plus = vitD_mean + vitD_std
vitD_std_minus = vitD_mean - vitD_std
```

```python
# Plotting 
fig, ax = plt.subplots(figsize=(10,6))
sns.histplot(data=df, x= df['VitD_levels'], kde='true', ax = ax);
ax.axvline(vitD_mean, color='red', label='mean');
ax.axvline(vitD_median, ls= '--', color='green', label='median');
ax.axvline(vitD_std_plus, color='black', label='+1 std');
ax.axvline(vitD_std_minus, color='black', label='-1 std');
ax.axvspan(vitD_std_plus, vitD_std_minus, color='yellow');
ax.legend();
ax.set_title('Distribution of Vitamin D levels')

```
